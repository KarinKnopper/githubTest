## Python

Som beskrevet tidligere går python og dataanalyse hånd i hånd. Meget af det man kan lave i Power BI ved hjælp af drag'n'drop, kan man også lave med pythonkode. Det kræver, at man importerer nogle biblioteker og værktøjer, hvilket er meget nemt når man bruger Jupyter Notebook. Nedenunder har jeg hentet tre csv filer og lavet den samme visualisering (næsten) som i Power BI, men der er lige nogle ting, der skal arbejdes på...

Jeg havde problemer med den ene csv fil da jeg hverken kunne bruge komma eller kolon til at separere, så jeg brugte tabulator, men den blev læst forkert ind, så jeg havde kun en kolonne med et meget langt navn og sære værdier. Det krævede nogle forsøg at få det til at fungere.

Der er 3 måder at hente filer på. Den ene er fra et lokalt drev, den anden er fra google drive og den tredje er via github. Det var klart nemmest fra github, når andre også skal have adgang til dataene.

Jeg startede med at importere pandas, så jeg kunne opbygge en dataframe. Der er et standard alias -pd, og en indbygget funktion til at læse filerne -pd.read_csv og det gemte jeg i et dataframe variabel med linket til filen på github som parameter. I det første dataframe -sol- indlæser jeg kun de første fem linjer, så jeg kan se hvad kolonnerne hedder.

```python
import pandas as pd

sol = pd.read_csv('https://raw.githubusercontent.com/KarinKnopper/githubTest/main/Solskinstimer.csv')
sol.head()
```

Det næste dataframe -vind- printer jeg helt.

```python
vind = pd.read_csv('https://raw.githubusercontent.com/KarinKnopper/githubTest/main/Vinddata.csv')

print(vind)
```

Den sidste fil er den med priserne. Det er den jeg har kæmpet lidt med.

```python
priser = pd.read_csv('https://raw.githubusercontent.com/KarinKnopper/githubTest/main/Elpriser.csv')
priser.head()
```

For at kunne skabe visualiseringer skal jeg bruge matplotlib, men da jeg kun skal bruge pyplot nøjes jeg med at importere det og giver det et alias. Mange af disse biblioteker har et standard alias, så man slipper for at skrive de til tider meget lange navne..
Her kan jeg så lave forskellige visualiseringer som bar chart, line chart, scatterplot og histogrammer.
Man kan navngive akserne, lave overskrift, ændre farve, skrifttype og størrelse og en ufattelig masse andre sjove ting.

```python
from matplotlib import pyplot as plt


plt.bar(priser.Dato, priser.Total, label="Elpriser")
plt.plot(sol.DateTime, sol['Sum of Solskin'], label="Solskinstimer", color='red')
plt.plot(vind['Sum of Middel vindhastighed'], label="Vind", color='green')
plt.xlabel("Dato")
plt.ylabel("Pris")
plt.title("Prissammenligning", fontsize=20, color="green")
plt.text(30, 55, "Meget sol")
plt.text(68, 12, "Meget vind")
plt.legend()
plt.show()
```
